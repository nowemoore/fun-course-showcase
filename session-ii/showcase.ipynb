## Control Flows Showcase

### Conditionals

#### If-Elif-Else

```python
time_of_day = "morning"

if time_of_day == "morning":
    print("I make coffee.")
elif time_of_day == "afternoon":
    print("I make tea.")
else:
    print("I make hot chocolate.")
```

#### Match-Case

```python
day = "Monday"

match day:
    case "Monday":
        print("I make coffee.")
    case "Tuesday":
        print("I make tea.")
    case "Wednesday":
        print("I make smoothie.")
    case "Thursday":
        print("I make juice.")
    case "Friday":
        print("I make hot chocolate.")
    case "Saturday" | "Sunday":
        print("I sleep in!")
    case _:
        print("That's not a day!")
```

### Loops

```python
books = ["Harry Potter", "Lord of the Rings", "1984", "To Kill a Mockingbird"]

for book in books:
    print(f"I read {book}.")
```

### While

```python
pages_left = 100

while pages_left > 0:
    print(f"Still reading... {pages_left} pages left.")
    pages_left -= 20
    
print("Book finished!")
```

## Functions

### Define a Function

```python
def make_coffee(coffee_type, sugar, milk):
    print(f"Making {coffee_type}...")
    
    if sugar:
        print("Adding sugar...")
    
    if milk:
        print("Adding milk...")
    
    return "Coffee is ready!"
```

### Call a Function

```python
make_coffee("espresso", sugar=True, milk=False)
```
